```python
# import all the things

%matplotlib inline
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
from matplotlib import gridspec
from matplotlib.finance import quotes_historical_yahoo_ochl
import datetime
import csv
import os
```

```python
# retrieve symbol list
with open('data/symbols.txt', 'r') as f:
    symbols = f.read().splitlines()

len(symbols), 'symbols listed'

# acquire data
quotes = []
for symbol in symbols:
    fname = 'data/' + symbol + ".csv"
    if os.path.isfile(fname):
        data = np.loadtxt(fname, delimiter=',')
        quotes.append(data)
        
len(quotes), 'data streams retrieved'
```

```python
# set up plot
f = plt.figure(figsize=(200,20))

gs = gridspec.GridSpec(4, 1)
px = plt.subplot(gs[0:3,0])

px.set_autoscaley_on(False)
px.set_ylim([0.5,1.5])

# prepare data arrays + draw
for data in quotes:
    dates = array([q[0] for q in data])
    opens = array([q[2] for q in data])
    closes = array([q[2] for q in data])
    highs = array([q[3] for q in data])
    lows = array([q[4] for q in data])
    volumes = array([q[5] for q in data])

    try:
        px.plot_date(dates, closes/closes, '-', alpha=0.0)
    except:
        True
        
    order = 8
    try:
        cmax = argrelextrema(data=highs, comparator=np.greater, order=order)
        px.plot(dates[cmax], highs[cmax]/closes[cmax], "k.", alpha=0.1)
    except:
        True
    try:
        cmin = argrelextrema(data=lows, comparator=np.less, order=order)
        px.plot(dates[cmin], lows[cmin]/closes[cmin], "k.", alpha=0.1)
    except:
        True

px.grid(True)
plt.show()
```
